```python
print("\nChoose Difficulty")
print("1. Easy")
print("2. Medium")
print("3. Hard")

choice = input("Enter choice (1-3): ")

if choice == "1":
    difficulty = "Easy"
    lives = 8
    reward = 10

elif choice == "2":
    difficulty = "Medium"
    lives = 6
    reward = 20

else:
    difficulty = "Hard"
    lives = 4
    reward = 30

print("\nDifficulty:", difficulty)
print("Lives:", lives)
```

```python
print("\nCategories Available")

category_names = list(categories.keys())

for i in range(len(category_names)):
    print(i+1, ".", category_names[i])

category_choice = int(input("\nSelect category: "))

category = category_names[category_choice-1]

print("\nCategory selected:", category)
```

```python
word = random.choice(list(categories[category].keys()))

hint = categories[category][word]

```

```python
score = 0

guessed_letters = []

wrong_letters = []

correct_guesses = 0

wrong_guesses = 0

hint_used = False

game_over = False
print(score)
print(guessed_letters)
```

```python
print("\nStarting Game...\n")

print("Category :", category)

print("Lives :", lives)

print("Score :", score)

print(stages[0])
```

```python
display_word = ""

for letter in word:
    display_word += "_ "

print("\nWord :",display_word)
```

```python
while lives > 0:

    # Display current word
    display_word = ""

    for letter in word:
        if letter in guessed_letters:
            display_word += letter + " "
        else:
            display_word += "_ "

    print("\nWord :", display_word)
    print("Wrong Letters :", wrong_letters)
    print("Lives :", lives)
    print("Score :", score)

    # Check win condition
    if "_" not in display_word:
        print("\n🏆 CONGRATULATIONS!")
        print("You guessed the word:", word)
        score += 50
        break

    choice = input("\nEnter a letter, 'hint', 'reveal', or entire word: ").lower()

    # Hint
    if choice == "hint":
        if not hint_used:
            print("💡 Hint:", hint)
            score -= 10
            hint_used = True
        else:
            print("Hint already used!")
        continue

    # Reveal a letter
    if choice == "reveal":
        if not reveal_used:

            hidden_letters = []

            for letter in word:
                if letter not in guessed_letters:
                    hidden_letters.append(letter)

            if hidden_letters:
                reveal_letter = random.choice(hidden_letters)
                guessed_letters.append(reveal_letter)
                score -= 20
                reveal_used = True

                print("🔓 Revealed Letter:", reveal_letter)

        else:
            print("Reveal already used!")

        continue

    # Guess entire word
    if len(choice) > 1:

        if choice == word:
            guessed_letters = list(word)
            score += 50

            print("\n🏆 CONGRATULATIONS!")
            print("You guessed the word:", word)

            break

        else:
            print("❌ Wrong word!")
            lives -= 2
            score -= 10

            if lives <= 0:
                break

            continue

    # Correct letter
    if choice in word:

        if choice not in guessed_letters:

            guessed_letters.append(choice)

            score += reward

            correct_guesses += 1

            print("✅ Correct!")

        else:
            print("Already guessed!")

    # Wrong letter
    else:

        if choice not in wrong_letters:

            wrong_letters.append(choice)

            lives -= 1

            score -= 5

            wrong_guesses += 1

            print("❌ Wrong Guess!")

            stage_index = min(6, len(stages)-1, 8-lives)
            print(stages[stage_index])

        else:
            print("Already guessed!")

# Game Over
if lives <= 0:

    print("\n💀 GAME OVER")
    print("The word was:", word)

# Statistics
total_attempts = correct_guesses + wrong_guesses

if total_attempts > 0:
    accuracy = (correct_guesses / total_attempts) * 100
else:
    accuracy = 0

print("\n========== GAME STATISTICS ==========")
print("Correct Guesses :", correct_guesses)
print("Wrong Guesses   :", wrong_guesses)
print("Accuracy        : {:.2f}%".format(accuracy))
print("Final Score     :", score)
```
