```python
import random
import csv
from datetime import datetime
today = datetime.now()
portfolio = {}
currency_symbols = {
    "1": ("USD", "$"),
    "2": ("INR", "₹"),
    "3": ("EUR", "€"),
    "4": ("GBP", "£")
}

print("✅ Libraries Imported Successfully!")
```

```python
# ==========================================
# 🏦 WELCOME SCREEN
# ==========================================

print("\n")
print("╔══════════════════════════════════════════════════════════════╗")
print("║               📈 STOCK PORTFOLIO TRACKER 📈                 ║")
print("╚══════════════════════════════════════════════════════════════╝")

name = input("\n👤 Enter Investor Name : ").title()

print("\n📅 Date :", today.strftime("%d-%m-%Y"))
print("⏰ Time :", today.strftime("%I:%M:%S %p"))

# ==========================================
# 💱 CURRENCY SELECTION
# ==========================================

print("\nChoose Currency")
print("1. USD ($)")
print("2. INR (₹)")
print("3. EUR (€)")
print("4. GBP (£)")

choice = input("\nEnter Choice (1-4): ")

if choice not in currency_symbols:
    choice = "1"

currency_name, symbol = currency_symbols[choice]

print(f"\n✅ Currency Selected : {currency_name} ({symbol})")

# ==========================================
# 📈 STOCK DATABASE (20 Stocks)
# ==========================================

stock_prices = {

    "AAPL":180,
    "TSLA":250,
    "GOOGL":140,
    "MSFT":330,
    "AMZN":135,

    "NVDA":450,
    "META":290,
    "NFLX":420,
    "IBM":165,
    "INTC":95,

    "ORCL":150,
    "AMD":175,
    "SAP":160,
    "SONY":105,
    "TCS":3800,

    "INFY":1650,
    "WIPRO":520,
    "HCL":1450,
    "RELIANCE":2900,
    "TATAMOTORS":720

}

# ==========================================
# 📋 DISPLAY STOCK LIST
# ==========================================

print("\n")
print("╔════════════════════════════════════════════╗")
print("║            📈 AVAILABLE STOCKS            ║")
print("╚════════════════════════════════════════════╝")

print("{:<15}{:<10}".format("SYMBOL", "PRICE"))
print("-"*30)

for stock, price in stock_prices.items():
    print("{:<15}{}{}".format(stock, symbol, price))

print("-"*30)

print("\n✅ Total Stocks Available :", len(stock_prices))
print("👤 Investor :", name)
print("💱 Currency :", currency_name)
```

```python

def add_stock():

    while True:

        print("\nAvailable Stocks")
        print("-"*35)

        for stock, price in stock_prices.items():
            print("{:<15}{}{}".format(stock, symbol, price))

        print("-"*35)

        stock = input("\nEnter Stock Symbol : ").upper()

        if stock not in stock_prices:
            print("❌ Invalid Stock Symbol!")
            continue

        try:
            quantity = int(input("Enter Quantity : "))

            if quantity <= 0:
                print("❌ Quantity must be greater than 0")
                continue

        except:
            print("❌ Invalid Quantity")
            continue

        if stock in portfolio:
            portfolio[stock] += quantity
        else:
            portfolio[stock] = quantity

        print(f"\n✅ {quantity} shares of {stock} added successfully!")

        again = input("\nAdd Another Stock? (yes/no): ").lower()

        if again != "yes":
            break
def remove_stock():

    if len(portfolio) == 0:
        print("\n Portfolio Empty!")
        return

    stock = input("\nEnter Stock Symbol to Remove : ").upper()

    if stock in portfolio:

        del portfolio[stock]

        print("✅ Stock Removed Successfully!")

    else:

        print("❌ Stock Not Found!")


# Update Quantity
def update_stock():

    if len(portfolio) == 0:
        print("\n❌ Portfolio Empty!")
        return

    stock = input("\nEnter Stock Symbol : ").upper()

    if stock not in portfolio:

        print("❌ Stock Not Found!")

        return

    print("Current Quantity :", portfolio[stock])

    try:

        qty = int(input("Enter New Quantity : "))

        if qty <= 0:
            print("❌ Quantity should be greater than 0")
            return

        portfolio[stock] = qty

        print("✅ Quantity Updated!")

    except:

        print("❌ Invalid Quantity")


# Search Stock
def search_stock():

    stock = input("\nEnter Stock Symbol : ").upper()

    if stock in stock_prices:

        print("\nStock Found")

        print("Symbol :", stock)

        print("Price :", symbol, stock_prices[stock])

    else:

        print("❌ Stock Not Available!")


# View Portfolio
def view_portfolio():

    if len(portfolio) == 0:

        print("\n📂 Portfolio Empty!")

        return

    print("\n")
    print("📊 YOUR PORTFOLIO")
    print("-"*65)

    print("{:<15}{:<12}{:<15}{:<15}".format(
        "Stock","Quantity","Price","Investment"))

    print("-"*65)

    total = 0

    for stock, qty in portfolio.items():

        investment = qty * stock_prices[stock]

        total += investment

        print("{:<15}{:<12}{:<15}{:<15}".format(
            stock,
            qty,
            symbol+str(stock_prices[stock]),
            symbol+str(investment)))

    print("-"*65)

    print("Total Investment :", symbol, total)


# Calculate Investment
def calculate_investment():

    if len(portfolio) == 0:

        print("\n❌ Portfolio Empty!")

        return

    total = 0

    highest_stock = ""
    highest_value = 0

    lowest_stock = ""
    lowest_value = 999999999

    for stock, qty in portfolio.items():

        investment = qty * stock_prices[stock]

        total += investment

        if investment > highest_value:

            highest_value = investment
            highest_stock = stock

        if investment < lowest_value:

            lowest_value = investment
            lowest_stock = stock

    print("\n💰 TOTAL INVESTMENT :", symbol, total)

    print("📈 Highest Investment")
    print(highest_stock, "-", symbol, highest_value)

    print("📉 Lowest Investment")
    print(lowest_stock, "-", symbol, lowest_value)


# Save TXT
def save_txt():

    file = open("Portfolio_Report.txt","w")

    file.write("STOCK PORTFOLIO REPORT\n")
    file.write("-"*40+"\n")

    file.write("Investor : "+name+"\n")
    file.write("Currency : "+currency_name+"\n")
    file.write("Date : "+today.strftime("%d-%m-%Y")+"\n\n")

    total = 0

    for stock, qty in portfolio.items():

        investment = qty*stock_prices[stock]

        total += investment

        file.write(f"{stock}  Qty:{qty}  Investment:{symbol}{investment}\n")

    file.write("\n")
    file.write("Total Investment : "+symbol+str(total))

    file.close()

    print("\n✅ Portfolio saved as Portfolio_Report.txt")


# Save CSV
def save_csv():

    file = open("Portfolio_Report.csv","w",newline="")

    writer = csv.writer(file)

    writer.writerow(["Stock","Quantity","Price","Investment"])

    total = 0

    for stock, qty in portfolio.items():

        investment = qty*stock_prices[stock]

        total += investment

        writer.writerow([stock,qty,stock_prices[stock],investment])

    writer.writerow([])

    writer.writerow(["","","Total",total])

    file.close()

    print("\n✅ Portfolio saved as Portfolio_Report.csv")


# Portfolio Summary
def portfolio_summary():

    if len(portfolio)==0:

        print("\nPortfolio Empty!")

        return

    total = 0

    for stock, qty in portfolio.items():

        total += qty*stock_prices[stock]

    print("\n⭐ PORTFOLIO SUMMARY")
    print("-"*35)

    print("Investor :", name)

    print("Stocks Owned :", len(portfolio))

    print("Currency :", currency_name)

    print("Portfolio Value :", symbol, total)
# Top 3 Investments
def top3_investments():

    if len(portfolio) == 0:
        print("\n❌ Portfolio is Empty!")
        return

    investment_list = []

    for stock, qty in portfolio.items():

        investment = qty * stock_prices[stock]

        investment_list.append((stock, investment))

    investment_list.sort(key=lambda x: x[1], reverse=True)

    print("\n🏆 TOP 3 INVESTMENTS")
    print("-"*40)

    for i, (stock, investment) in enumerate(investment_list[:3], start=1):

        print(f"{i}. {stock:<15}{symbol}{investment}")
# Profit / Loss Calculator
def profit_loss():

    if len(portfolio) == 0:
        print("\n❌ Portfolio is Empty!")
        return

    stock = input("\nEnter Stock Symbol : ").upper()

    if stock not in portfolio:

        print("❌ Stock not found in Portfolio!")

        return

    purchase_price = stock_prices[stock]

    current_price = float(input("Enter Current Market Price : "))

    qty = portfolio[stock]

    profit = (current_price - purchase_price) * qty

    print("\n📊 PROFIT / LOSS REPORT")
    print("-"*40)

    print("Stock :", stock)
    print("Purchase Price :", symbol, purchase_price)
    print("Current Price  :", symbol, current_price)
    print("Quantity :", qty)

    if profit > 0:

        print(f"🟢 Profit : {symbol}{profit:.2f}")

    elif profit < 0:

        print(f"🔴 Loss : {symbol}{abs(profit):.2f}")

    else:

        print("⚪ No Profit No Loss")
```

```python
while True:

    print("\n" + "="*60)
    print("📈 STOCK PORTFOLIO TRACKER - MAIN MENU")
    print("="*60)

    print("1. ➕ Add Stock")
    print("2. ❌ Remove Stock")
    print("3. ✏️ Update Quantity")
    print("4. 🔍 Search Stock")
    print("5. 📊 View Portfolio")
    print("6. 💰 Calculate Investment")
    print("7. 📄 Save Portfolio as TXT")
    print("8. 📁 Save Portfolio as CSV")
    print("9. 🏆 Top 3 Investments")
    print("10. 💹 Profit / Loss Calculator")
    print("11. ⭐ Portfolio Summar")
    print("12. 🚪 Exit")

    choice = input("\nEnter your choice (1-10): ")

    if choice == "1":
        add_stock()

    elif choice == "2":
        remove_stock()

    elif choice == "3":
        update_stock()

    elif choice == "4":
        search_stock()

    elif choice == "5":
        view_portfolio()

    elif choice == "6":
        calculate_investment()

    elif choice == "7":
        save_txt()

    elif choice == "8":
        save_csv()

    elif choice == "9":
        top3_investments()

    elif choice == "10":
        profit_loss()

    elif choice == "11":
        portfolio_summary()

    elif choice == "12":


        print("\n" + "🌟"*20)
        print("      THANK YOU FOR USING")
        print("   STOCK PORTFOLIO TRACKER")
        print("🌟"*20)

        print(f"\n👤 Investor : {name}")
        print("💼 Happy Investing!")
        print("🚀 Have a Great Day!")

        break

    else:
        print("\n❌ Invalid Choice! Please select from 1 to 10.")
```
