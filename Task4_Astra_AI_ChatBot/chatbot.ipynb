```python
# ============================================================
# 🤖 CHATBOT FUNCTIONS
# ============================================================

# Help Menu
def show_help():

    print("\n📚 AVAILABLE COMMANDS")
    print("-"*45)

    commands = [
        "hello / hi",
        "how are you",
        "your name",
        "who developed you",
        "python",
        "ai",
        "machine learning",
        "deep learning",
        "gen ai",
        "chatbot",
        "horizon techx",
        "date",
        "time",
        "joke",
        "quote",
        "fact",
        "help",
        "exit"
    ]

    for command in commands:
        print("👉", command)


# Save Chat History
def save_history():

    with open("Chat_History.txt", "w") as file:

        file.write("========== CHAT HISTORY ==========\n\n")

        for line in chat_history:
            file.write(line + "\n")

    print("\n💾 Chat History Saved Successfully!")


# Greeting
def greet():

    greetings = [

        f"Hello {user_name}! 👋",

        f"Hi {user_name}! 😊",

        f"Welcome {user_name}! 🤖",

        f"Nice to see you {user_name}! 🌸"

    ]

    return random.choice(greetings)


# Joke
def tell_joke():

    return random.choice(jokes)


# Motivation
def motivation():

    return random.choice(quotes)


# Fun Fact
def fun_fact():

    return random.choice(facts)


# Current Time
def current_time():

    return datetime.now().strftime("%I:%M:%S %p")


# Current Date
def current_date():

    return datetime.now().strftime("%d-%m-%Y")


# ChatBot Brain
def chatbot_response(message):

    global question_count

    question_count += 1

    message = message.lower().strip()

    # Greetings
    if message in ["hi", "hello", "hey"]:

        return greet()

    # How are you
    elif "how are you" in message:

        return "😊 I'm doing great! Thank you for asking."

    # Name
    elif "your name" in message:

        return f"My name is {bot_name}. 🤖"

    # Developer
    elif "who developed you" in message:

        return f"I was developed by {developer}."

    # Time
    elif message == "time":

        return "⏰ Current Time : " + current_time()

    # Date
    elif message == "date":

        return "📅 Today's Date : " + current_date()

    # Joke
    elif message == "joke":

        return "😂 " + tell_joke()

    # Motivation
    elif message == "quote":

        return "💡 " + motivation()

    # Fact
    elif message == "fact":

        return "📚 " + fun_fact()

    # Help
    elif message == "help":

        show_help()

        return None

    # Knowledge Base
    elif message in knowledge:

        return knowledge[message]

    # Thank You
    elif "thank" in message:

        return "😊 You're Welcome!"

    # Goodbye
    elif message == "bye":

        return "👋 Goodbye! Type 'exit' to close the chatbot."

    # Unknown
    else:

        return "🤔 Sorry! I don't know the answer to that. Type 'help' to see available commands."
```
